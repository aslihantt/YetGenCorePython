# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

class Bus(Vehicle):
    def seating_capacity(self, capacity=50):
        return capacity
    
my_bus = Bus("Mercedes", "Travego", 2020)
print(my_bus.seating_capacity()) # 50
````

School_bus'ın aynı zamanda Vehicle sınıfının bir örneği olup olmadığını belirleyiniz.

````python
# 2.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def seating_capacity(self, capacity=50):
        return capacity

class School_bus(Vehicle):
    pass

my_school_bus = School_bus("Mercedes", "Travego", 2022)
print(my_school_bus.seating_capacity()) # 50
````
